```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import pylab
import pandas as pd
import json
print(plt.style.available)
```

```python
plt.style.use('seaborn-paper')

from matplotlib import rc
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
#rc('font',**{'family':'sans-serif','sans-serif':['Computer Modern Sans serif']})
#rc('font',**{'family':'serif','serif':['Times']})
rc('font',**{'family':'serif'})
rc('text', usetex=True)

SMALL_SIZE = 14
MEDIUM_SIZE = 18
BIGGER_SIZE = 20

rc('font', size=SMALL_SIZE)          # controls default text sizes
rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
list_of_files =  [('10AA_OB_ff94_3/fitnesses.dat', 'OBff94'),
                  ('10AA_OB_ff94_1/fitnesses.dat', 'OBff94 + basilisk, 1'),
                  ('10AA_OB_ff94_2/fitnesses.dat', 'OBff94 + basilisk, 2'),
                  ('10AA_amber_7/fitnesses.dat', 'gaff, ncyc=2000'),
                  ('10AA_amber_4/fitnesses.dat', 'gaff + basilisk, ncyc=2000, 1'),
                  ('10AA_amber_6/fitnesses.dat', 'gaff + basilisk, ncyc=2000, 2'),
                  ('10AA_amber_8/fitnesses.dat', 'gaff, ncyc=500'),
                  ('10AA_amber_5/fitnesses.dat', 'gaff + basilisk, ncyc=500')
                  ]

fitnesses = []
labels = []
#Read JSON data into the datastore variable
for i, (filename, label) in enumerate(list_of_files):
    with open(filename, 'r') as f:
        labels.append(label)
        print(labels[i])
        fitnesses.append(np.array(json.load(f)))
        print(fitnesses[i])

list_of_files2 = [('10AA_OB_ff94_3/initial_fitness.dat', 'OBff94'),
                  ('10AA_OB_ff94_1/initial_fitness.dat', 'basilisk + OBff94, 1'),
                  ('10AA_OB_ff94_2/initial_fitness.dat', 'basilisk + OBff94, 2'),
                  ('10AA_amber_7/initial_fitness.dat', 'gaff, ncyc=2000'),
                  ('10AA_amber_4/initial_fitness.dat', 'basilisk + gaff, ncyc=2000, 1'),
                  ('10AA_amber_6/initial_fitness.dat', 'basilisk + gaff, ncyc=2000, 2'),
                  ('10AA_amber_8/initial_fitness.dat', 'gaff, ncyc=500'),
                  ('10AA_amber_5/initial_fitness.dat', 'basilisk + gaff, ncyc=500')
                  ]

initial_fitnesses = []
labels2 = []
#Read JSON data into the datastore variable
for i, (filename, label) in enumerate(list_of_files2):
    with open(filename, 'r') as f:
        labels2.append(label)
        print(labels2[i])
        initial_fitnesses.append(np.array(json.load(f)))
        print(initial_fitnesses[i])
```

```python
plt.figure(num=None, figsize=(8, 6))
for i in np.arange(0, len(fitnesses)-5):
    plt.plot(fitnesses[i], label=labels[i], marker='.', markersize=8, linewidth=1.6)

plt.legend()
plt.title(r'\textbf{Energy convergence}')
plt.xlabel("Iteration number")
plt.ylabel("Energy [kcal/mol]")

plt.figure(num=None, figsize=(8, 6))
for i in np.arange(3, len(fitnesses)):
    plt.plot(fitnesses[i], label=labels[i], marker='.', markersize=8, linewidth=1.6)

plt.legend()
plt.title(r'\textbf{Energy convergence}')
plt.xlabel("Iteration number")
plt.ylabel("Energy [kcal/mol]")
```

```python
plt.figure(num=None, figsize=(8, 6))
for i in np.arange(0, len(initial_fitnesses)):
    plt.plot(fitnesses[i]-initial_fitnesses[i], label=labels[i], marker='.', markersize=8, linewidth=1.6)

plt.legend()
plt.title(r'\textbf{Energy convergence}')
plt.xlabel("Iteration number")
plt.ylabel("$E - E_{0}$ [kcal/mol]")
```
